```python
import scipy.signal as signal
import numpy as np
import matplotlib.pyplot as plt

signals = data['signals']

sampling_rate = 1000
window_length = context.params['spectrogram_params']['window_length']
shift = context.params['spectrogram_params']['shift']
logarithm = context.params['spectrogram_params']['logarithm']

factor = int(sampling_rate/1000)
window_length = window_length * factor
shift = shift * factor

fs = 1e3 * factor

noverlap = window_length - shift
nperseg = window_length

f, t, sxx = signal.spectrogram(signals, fs=fs, noverlap=noverlap, nperseg=nperseg, axis=0)

sxx = np.moveaxis(sxx, -2, -1)
```
